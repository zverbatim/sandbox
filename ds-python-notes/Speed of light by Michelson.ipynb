```python
mu = speed.mean()
sigma = speed.std()
size = len(speed)
normal = np.sort(np.random.normal(mu, sigma, size=size)) 

plt.plot(np.sort(speed), marker='.', linestyle='none', color='b', label='actual measurements')
plt.plot(normal, marker='.', color='g', label='normal distribution using observed params', alpha=0.5)
plt.legend()
plt.show()
```

```python
def bootsrap_replicate(list, fx):
    """
        Apply a function (np.mean, np.std) to a newly sampled list with replacement
        Used to make large simulated mesurements (of the applied) function for determining the 
        normal distribution of an event
    """
    bootstrap_list = np.random.choice(list, len(list))
    return fx(bootstrap_list)
```

```python
bootsrap_replicate(speed, np.mean)
```

```python
bootsrap_replicate(speed, np.std)
```

```python
def replicates(list, s, fx):
    """
        Generate large number of replicates
    """
    r = np.empty(s)
    for i in range(s):
        r[i] = bootsrap_replicate(list, fx)
    return r
```

```python
r = replicates(speed, len(speed), np.mean)
```

```python
plt.plot(np.sort(r), marker='.', linestyle='none', color='b', label='bootstrap means')
plt.legend()
plt.show()
```

```python
#  confidence interval
np.percentile(r, [25, 50, 75])
```
