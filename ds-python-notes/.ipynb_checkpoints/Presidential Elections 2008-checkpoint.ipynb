```python
dem_fl = swing_states.loc['FL']['dem_share'].values
dem_oh = swing_states.loc['OH']['dem_share'].values
dem_pa = swing_states.loc['PA']['dem_share'].values
```

```python
def ecdf(list):
    """
      Empirical cumulative distribution function
      Afterwards can be called
      plt.plot(x, y, marker='.')
    """
    x = np.sort(list)
    y = np.arange(1, len(list) + 1) / len(list)
    return x, y
```

```python
x_fl, y_fl = ecdf(dem_fl)
x_oh, y_oh = ecdf(dem_oh)
x_pa, y_pa = ecdf(dem_pa)
```

```python
plt.plot(x_fl, y_fl, marker='.', linestyle='none', color='g')
plt.plot(x_pa, y_pa, marker='.', linestyle='none', color='r')
plt.plot(x_oh, y_oh, marker='.', linestyle='none', color='b')
plt.xlabel('dem_share')
plt.ylabel('distribution')
plt.show()
```

```python
sns.boxplot(y=swing_states.index, x = 'dem_share', data=swing_states)
plt.ylabel('state')
plt.xlabel('dem_share')
plt.show()
```

```python
# look for correlations
total = swing_states['total_votes']
dem_share= swing_states['dem_share']
plt.plot(total, dem_share, marker='.', linestyle='none')
plt.show()
```

```python
plt.plot(swing_states.loc['FL']['total_votes'], dem_fl, marker='.', linestyle='none', color='r')
plt.show()

plt.plot(swing_states.loc['OH']['total_votes'], dem_oh, marker='.', linestyle='none', color='g')
plt.show()

plt.plot(swing_states.loc['PA']['total_votes'], dem_pa, marker='.', linestyle='none')
plt.show()
```

```python
# fitting a linear model for x -> numbers of votes, y = dem_share
slope,intercept = np.polyfit(total, dem_share,1)
print(slope, intercept)
```

```python
model_in = np.array([0,total.max()])
model_out = model_in * slope + intercept
```

```python
plt.plot(model_in, model_out)
plt.plot(total, dem_share, marker='.', linestyle='none')
plt.margins(0.02)
plt.show()
```

```python

for _ in range(50):
    i = np.arange(len(total))

    # shufle them
    i = np.random.choice (i, len(i))

    # sample
    t = total[i]
    ds = dem_share[i]

    # get the slope and intercept
    x, y = np.polyfit(t , ds, 1)
    
    # fro the plot
    m_in = np.array([0,total.max()])
    m_out = m_in * x + y

    # use a low alpha for nice visual
    plt.plot(m_in, m_out, alpha=0.1, color='g')    
    
plt.plot(total, dem_share, marker='.', linestyle='none')
plt.margins(0.02)
plt.show()
```
